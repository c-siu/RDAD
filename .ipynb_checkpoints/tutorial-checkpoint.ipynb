```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import gudhi as gd
import RDAD
```

```python
# The package tqdm is optional
_has_tqdm = True
try: import tqdm
except ImportError: _has_tqdm = False
```

## Data Generation and Parameter Selection

We first generate the dataset. We will generate the "David and Goliath" dataset, which conists of points densely sampled from a small square, and points sparsely sampled on a big square. The RDAD filtration weights distances by the local density, and hence the persistence of the smaller but denser square annulus under the RDAD filtration is prolonged.

```python
# generate data points
def generate_square_annulus_sample(R, r, N, seed = 0):
    # generate a uniform sample on a square annulus by the acceptance-rejection method
    
    # The square annulus is centered at (0, 0).
    # The NE and SW vertices of the outer square are (R, R) and (-R, R).
    # Those of the inner square are (r, r) and (-r, r).
    # The sample size is N.
    # "Seed" is the seed for the random number generator.
    
    # The output is an Nx2 numpy array, each row of which is a sample point.
    
    # The acceptance-rejection method samples points on the big square and 
    # discard points on the small squares.
    # This is done iteratively until N points have been sampled.
    # Since we sample from the big square more than N points at each iterations
    # in such a way that on average, only 1 iteration is needed.
    
    rng = np.random.default_rng(seed = seed)
    
    success_prob = 1 - r**2/R**2 # prob of not falling on the smaller square
    N_plus = int(N/success_prob*1.1) # num of pts sampled on the bigger square at each iteration
    n = 0 # cumulative number of points on annulus
    points = [] # list of sample points from each iterations
    while n < N:
        new_points = R * rng.uniform(-1, 1, size = (N_plus, 2))
        on_annulus = np.logical_or(abs(new_points[:, 0]) > r, abs(new_points[:, 1]) > r)
        points.append(new_points[on_annulus])
        n += sum(on_annulus)
    points = np.vstack(points)[:N, :]
    
    return points

R1, r1, N1 = 1.1, 1, 200
R2, r2, N2 = .12, .1, 300
x0 = 4
points_goliath = generate_square_annulus_sample(R1, r1, N1, seed = 0)
points_david = generate_square_annulus_sample(R2, r2, N2, seed = 1) + np.array([x0, 0])
points = np.vstack([points_goliath, points_david])

fig, ax = plt.subplots()
ax.scatter(*points.T, s = 1)
ax.set_aspect('equal')
plt.show(fig)
```

We would like to compute the DTM filtration and RDAD filtration for this dataset, but before that, we must choose the parameters.

```python
sample_size = points.shape[0]
k_denest = int(np.ceil(np.log10(sample_size)) ** 2) # k for nearest neigbhor density estimation
m_DTM = 0.005 # mass threshold
dx = 0.02
xlim = [-1.3, 4.3]
ylim = [-1.3, 1.3]
```

In particular, we need to generate the grid on which the filtration functions will be evaluated. The grid is implemented as a Python dictionary. Its entries are used in the computation, and some of them will be useful for plotting the reusults.

```python
grid = RDAD.generate_grid(xlim = xlim, ylim = ylim, dx = dx, dy = dx)
```

## Computing the RDAD filtration

We compute the RDAD filtration first. We will compare it against the DTM, distance and DAD filtrations in subsequent sections.

The `RDAD` function takes the input points and the parameters, and return a 1D numpy array, which upon reshaping, is the function values of the mathematical function $RDAD$ on the grid. The reshaped array may be used to generate a contour plot of the filtration function.

```python
RDAD_filtration = RDAD.RDAD(points, grid, m_DTM, k_denest = k_denest)
```

```python
RDAD_filtration_reshaped = RDAD_filtration.reshape(grid["Xgrid"].shape)
fig, ax = plt.subplots()
ax.contour(grid["Xgrid"], grid["Ygrid"], RDAD_filtration_reshaped, levels = 30)
ax.set_aspect('equal')
ax.set_title('RDAD')
plt.show(fig)
```

Once the filtration function values have been computed, its topological features can be computed.

```python
RDAD_persistences, RDAD_pairs, RDAD_vals = RDAD.compute_persistence(grid, RDAD_filtration)
```

The first output contains the birth and death times of all topological features, and it can be passed to the `Gudhi` function `plot_persistence_diagram` to generate the persistence diagrams.

The second and third outputs contain the indices of creating and killing pixels, and the filtration function values . In the sections "Identification of Holes" and "Bootstrapping and Significant Features" below, we will use them to plot features corresponding to different points on persistence diagrams.

```python
fig, ax = plt.subplots()
gd.plot_persistence_diagram(
    RDAD_persistences,
    legend=True,
    axes=ax
    )
ax.set_aspect('equal')
ax.set_title('RDAD')
plt.show(fig)
```

## Computation of the DTM, DAD and Distance Filtrations 

We now compute the DTM, DAD and distance filtrations by supplying different parameters to the `RDAD` function.

For the DTM filtration, it computation can be done with the `RDAD` function in the repo as well by setting the variable `weighted_flag` to be `False`.

```python
DTM_filtration = RDAD.RDAD(points, grid, m_DTM, weighted_flag = False)
```

```python
DTM_persistences, DTM_pairs, DTM_vals = RDAD.compute_persistence(grid, DTM_filtration)
```

```python
def plot_filtration_contour_and_persistence_diagrams(grid, filtration, persistences):
    
    #fil_on_grid = filtration.reshape(grid.Xgrid.shape)
    fil_on_grid = filtration.reshape(grid["Xgrid"].shape)
    fig, axes = plt.subplots(1, 2)
    axes[0].contour(grid["Xgrid"], grid["Ygrid"], fil_on_grid, levels = 30)
    
    gd.plot_persistence_diagram(
        persistences,
        legend=True,
        axes=axes[1]
        )
    axes[1].set_title("")

    for ax in axes: ax.set_aspect('equal')

    return fig, axes
```

```python
fig, axes = plot_filtration_contour_and_persistence_diagrams(grid, DTM_filtration, DTM_persistences)
fig.set_figheight(4)
fig.set_figwidth(12)
fig.suptitle("DTM")
plt.show(fig)
```

We may also compute the distance filtration and DAD filtration by setting the variable `m_DTM` to be `0`. The resultant plots look similar to the case with positive `m_DTM` because this dataset is clean.

```python
dist_filtration = RDAD.RDAD(points, grid, m_DTM = 0, weighted_flag = False)
DAD_filtration = RDAD.RDAD(points, grid, m_DTM = 0, k_denest = k_denest)
dist_persistences, dist_pairs, dist_vals = RDAD.compute_persistence(grid, dist_filtration)
DAD_persistences, DAD_pairs, DAD_vals = RDAD.compute_persistence(grid, DAD_filtration)
```

```python
fig, axes = plot_filtration_contour_and_persistence_diagrams(grid, dist_filtration, dist_persistences)
fig.set_figheight(4)
fig.set_figwidth(12)
fig.suptitle("dist")
plt.show(fig)

fig, axes = plot_filtration_contour_and_persistence_diagrams(grid, DAD_filtration, DAD_persistences)
fig.set_figheight(4)
fig.set_figwidth(12)
fig.suptitle("DAD")
plt.show(fig)
```

## Identification of Holes

We may have a sense of the locations of the topological features on the persistence diagrams by investigating their killing pixels. Below, we consider the persistence diagrams of the RDAD filtration and the DTM filtration and find the most persistent 1-dimensional features under these filtrations by plotting their killing pixels.

Recall the second and third outputs of the `RDAD` function contain the indices of creating and killing pixels, and the filtration function values. We explain them in greater detail here. In the section "Bootstrapping and Significant Features", we discuss a one-line function to plot all significant features.

The second output, which we have defined to be `RDAD_pairs`, is a pair of two lists. The first list, `RDAD_pairs[0]` is a list of features that die in finite time; the second list, `RDAD_pairs[1]`, is a list of those that die at infinity. The `d`-th entry of each of `RDAD_pairs[0]` and `RDAD_pairs` is a numpy array of features of dimension `d`, whose `i`-th row is the index of the `i` creating, and killing (if applicable), cell in the lists `grid["Xgrid"]` and `grid["Ygrid"]`.

The third output has a similar structure, but instead of storing the indices of creating and killing cells, it stores the filtration values at the creating and killing cells, which are the birth and death times of the corresponding features

We find the killing pixel of the most persistent 1-dimensional feature under the RDAD filtration. It will be inside the smaller square. It is the most persistent under the RDAD filtration because the RDAD filtration "magnifies" dense features and the small square is very dense.

```python
# find the index of the most persistent 1-dimensional feature

mortal_vals = RDAD_vals[0]
mortal_1D_vals = mortal_vals[1]
persistence_mortal_1D_features = mortal_1D_vals[:, 1] - mortal_1D_vals[:, 0]
idx_most_persistent_mortal_1D_feature = np.argmax(persistence_mortal_1D_features)
print("The index of the most persistent 1D feature is {}.".format(idx_most_persistent_mortal_1D_feature))
```

```python
# find the coordinates of the killing pixel of the most persistent 1-dimensional feature

mortal_pairs = RDAD_pairs[0]
mortal_1D_pairs = mortal_pairs[1]
most_persistent_mortal_1D_pairs = mortal_1D_pairs[idx_most_persistent_mortal_1D_feature]
killing_pixel_idx = most_persistent_mortal_1D_pairs[1]
killing_pixel_xcoord = grid["Xgrid"].flatten()[killing_pixel_idx]
killing_pixel_ycoord = grid["Ygrid"].flatten()[killing_pixel_idx]

fig, ax = plt.subplots()
ax.scatter(*points.T, s = 1)
ax.set_aspect('equal')
ax.scatter([killing_pixel_xcoord], [killing_pixel_ycoord], marker = 'x')
ax.set_title('Killing Pixel of the Most Persistent 1D Feature under the RDAD Filtration')
plt.show(fig)
```

We do the same for the DTM filtration to see that the most persistent feature is the bigger square, because only size matters for DTM when the dataset is clean.

```python
idx_most_persistent_mortal_1D_feature = np.argmax(DTM_vals[0][1][:, 1] - DTM_vals[0][1][:, 0])
killing_pixel_idx = DTM_pairs[0][1][idx_most_persistent_mortal_1D_feature][1]
killing_pixel_xcoord = grid["Xgrid"].flatten()[killing_pixel_idx]
killing_pixel_ycoord = grid["Ygrid"].flatten()[killing_pixel_idx]

fig, ax = plt.subplots()
ax.scatter(*points.T, s = 1)
ax.set_aspect('equal')
ax.scatter([killing_pixel_xcoord], [killing_pixel_ycoord], marker = 'x')
ax.set_title('Killing Pixel of the Most Persistent 1D Feature under the DTM Filtration')
plt.show(fig)
```

## Bootstrapping and Plotting Significant Features


We now compute a confidence band by subsample bootstrapping to check that indeed there are two significant loops.

Again, we need to specify some parameters.
`B` is the number of bootstrap samples. It is 100 by default.
`alpha` is the empirical confidence level. It is 0.05 by default.
`bootstrap_dim` is the dimension of the topological features whose variability the confidence band captures. By default, it is 1, which means we are interested in the variability of loops.

For the purpose of this demonstration, we put $B = 10$ and $\alpha = 0.1$ to reduce runtime.
We also fix the seeds of the subsampling process to ensure reproducability. By default, the seeds themselves are generated randomly. When other seeds are used, the confidence band may be very wide because we have chosen a large $\alpha$ and a small $B$.
We keep `bootstrap_dim` as $1$. 

```python
bootstrap_dim = 1
# B, alpha = 100, 0.05
B, alpha = 10, 0.1
seeds = 2 + 2*np.arange(B).astype(int)
# seeds = None
```

```python
band = RDAD.compute_confidence_band(
    points, RDAD_persistences, grid, m_DTM, k_denest = k_denest, 
    bootstrap_dim = bootstrap_dim, B = B, alpha = alpha, seeds = seeds)
```

```python
fig, ax = plt.subplots()
gd.plot_persistence_diagram(
    RDAD_persistences,
    legend=True,
    axes=ax,
    band = band
    )
ax.set_aspect('equal')
ax.set_title('Persistence Diagrams of RDAD with a confidence band')
plt.show(fig)
```

We may also plot the killing pixels of the significant loops with finite death times. Below they are marked with crosses.

```python
killing_pixels = RDAD.get_significant_features(RDAD_pairs, RDAD_vals, grid, dim = 1, band = band)
```

```python
fig, ax = plt.subplots()
ax.scatter(*points.T, s = 1)
ax.set_aspect('equal')
ax.scatter(*killing_pixels.T, marker = 'x')
ax.set_title('Significant Features under the RDAD Filtration')
plt.show(fig)
```

We compare subsample bootstrapping with oracle bootstrapping. We generate 100 other samples with different seeds and compute the bottleneck distances between the resultant 1-dimensional persistent diagrams and the persistence diagram above. We take two times of the $\alpha^{th}$ quantile of the bottleneck distances as the confidence bound.

```python
def generate_alternative_Davids_and_Goliaths(R1, R2, r1, r2, N1, N2, x0, seed):
    seed_goliath = seed
    seed_david = seed
    if seed is not None:
        seed_david += 1
    points_goliath = generate_square_annulus_sample(R1, r1, N1, seed = seed_goliath)
    points_david = generate_square_annulus_sample(R2, r2, N2, seed = seed_david) + np.array([x0, 0])
    points = np.vstack([points_goliath, points_david])
    return points

RDAD_persistences_dim1 = [bd for (dim, bd) in RDAD_persistences if dim == bootstrap_dim]
bottleneck_distances = []
print("Computing persistence diagrams of oracle bootstrap samples.")
if _has_tqdm: iterator = tqdm.tqdm(range(B))
else: iterator = range(B)
for i in iterator:
    # alt_points = generate_alternative_Davids_and_Goliaths(R1, R2, r1, r2, N1, N2, x0, seeds[i])
    alt_points = generate_alternative_Davids_and_Goliaths(R1, R2, r1, r2, N1, N2, x0, None)
    alt_RDAD_filtration = RDAD.RDAD(alt_points, grid, m_DTM = m_DTM, k_denest = k_denest)
    alt_RDAD_persistences, alt_RDAD_pairs, alt_RDAD_vals = RDAD.compute_persistence(grid, alt_RDAD_filtration)

    alt_RDAD_persistences = [bd for (dim, bd) in alt_RDAD_persistences if dim == bootstrap_dim]
    bottleneck_distance = gd.bottleneck_distance(RDAD_persistences_dim1, alt_RDAD_persistences)
    bottleneck_distances.append(bottleneck_distance)

quan = np.quantile(
    bottleneck_distances,
    1 - alpha
    )

alt_band = 2*quan
print("Done computing persistence diagrams of oracle bootstrap samples and the confidence band.")
```

```python
fig, ax = plt.subplots()
gd.plot_persistence_diagram(
    RDAD_persistences_dim1 ,
    legend = False,
    axes=ax,
    colormap = matplotlib.cm.Set1.colors[1:]
    )
ax.axline([0, band], slope = 1, label = 'subsample', c = matplotlib.cm.Set1.colors[0])
ax.axline([0, alt_band], slope = 1, label = 'oracle', linestyle = 'dashed', c = matplotlib.cm.Set1.colors[2])
ax.legend()
ax.set_aspect('equal')
ax.set_title('Persistence Diagrams of RDAD with Confidence Bands')
plt.show(fig)
```
